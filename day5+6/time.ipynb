```python
df = yf.download("AAPL", start="2020-01-01", end="2024-12-31")
df = df[['Open','Close']] #just doing this for simplification
df.head()
```

```python
df['Close'].plot(title='Apple Stock Closing Price')
plt.show()
```

```python
result = seasonal_decompose(df['Close'], model='multiplicative', period=30)
result.plot()
plt.show()
```

```python
adf_result = adfuller(df['Close'].dropna())
print(f"ADF Statistic: {adf_result[0]}")
print(f"p-value: {adf_result[1]}")
```

Index 0: Test statistic

Index 1: p-value

there are also other critical values, number of lags used, etc. which are not printed here to keep it simple.
