```python
import matplotlib.pyplot as mp
import pandas as pd
import seaborn as sns

ds=sns.load_dataset('tips')
ds.head()
corr_matrix = ds.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
mp.title("Correlation Heatmap")
mp.show()
```
